```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the uploaded dataset
# Replace 'healthcare-dataset-stroke-data.csv' with your file's name if different
df = pd.read_csv('healthcare-dataset-stroke-data.csv')

# Data cleaning: Drop rows with missing values for simplicity
df = df.dropna()

# Question 1: Age distribution
plt.figure(figsize=(8, 5))
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.grid(True)
plt.show()
print("Q1: What is the age distribution of patients?")
print(f"The age distribution shows most patients are between {df['age'].min():.0f}–{df['age'].max():.0f} years, with a peak around {int(df['age'].mode()[0])} years.")

# Question 2: Glucose level by stroke status
plt.figure(figsize=(8, 5))
df.boxplot(column='avg_glucose_level', by='stroke', grid=False, patch_artist=True)
plt.title('Average Glucose Level by Stroke Status')
plt.suptitle('')
plt.xlabel('Stroke Status (0 = No, 1 = Yes)')
plt.ylabel('Average Glucose Level (mg/dL)')
plt.show()
glucose_stats = df.groupby('stroke')['avg_glucose_level'].describe()
print("\nQ2: How does average glucose level vary between patients with and without stroke?")
print(f"No stroke patients have a mean glucose level of {glucose_stats.loc[0, 'mean']:.1f} mg/dL (SD: {glucose_stats.loc[0, 'std']:.1f}).")
print(f"Stroke patients have a mean glucose level of {glucose_stats.loc[1, 'mean']:.1f} mg/dL (SD: {glucose_stats.loc[1, 'std']:.1f}).")

# Question 3: Correlation between BMI and stroke status
correlation = df['bmi'].corr(df['stroke'])
plt.figure(figsize=(8, 5))
plt.scatter(df['bmi'], df['stroke'], alpha=0.5, color='purple')
plt.title('BMI vs Stroke Status')
plt.xlabel('BMI')
plt.ylabel('Stroke Status (0 = No, 1 = Yes)')
plt.grid(True)
plt.show()
print("\nQ3: Is there a correlation between BMI and stroke status?")
print(f"The correlation coefficient between BMI and stroke status is {correlation:.2f}.")

# Question 4: Gender distribution among stroke patients
stroke_patients = df[df['stroke'] == 1]
gender_counts = stroke_patients['gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightpink'])
plt.title('Gender Distribution Among Stroke Patients')
plt.show()
print("\nQ4: What is the gender distribution among stroke patients?")
print(f"Among stroke patients, {gender_counts.get('Male', 0)/len(stroke_patients)*100:.1f}% are male, and {gender_counts.get('Female', 0)/len(stroke_patients)*100:.1f}% are female.")
```
